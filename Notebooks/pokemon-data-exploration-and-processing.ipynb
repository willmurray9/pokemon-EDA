# Data Munging: Filling Missing Values

```python
# check for missing values
raw_df.info()
```

```python
# start by looking at missing Type_2 values
raw_df[raw_df['Type_2'].isnull()]

# from my own knowledge, I can tell these values are not errors, these pokemon only have one type
```

```python
# fill in the nulls
raw_df['Type_2'].fillna('none', inplace = True)
```

```python
# check to confirm there are no longer any nulls
raw_df.Type_2.isnull().any()
```

```python
# look at the info again
raw_df.info()
```

```python
# next look at the nulls for Pr_Male
raw_df[raw_df['Pr_Male'].isnull()]
```

```python
# notice these nulls indicate a pokemon does not have a gender, so lets fill in the rows with that information
# however, we will want this column to remain a float, so lets instead fill the rows with the median value
raw_df['Pr_Male'].fillna((raw_df['Pr_Male'].median()), inplace = True)
```

```python
# check to confirm there are no longer any nulls
raw_df['Pr_Male'].isnull().any()
```

```python
raw_df.head()
```

```python
# look at the info again
raw_df.info()
```

```python
# finally, look at the missing values for Egg_Group_2
raw_df[raw_df['Egg_Group_2'].isnull()]
```

```python
# dig a little deeper to get a better understanding of the Egg Groups
raw_df['Egg_Group_1'].value_counts()
```

```python
# now look at Egg_Group_2 groups
raw_df['Egg_Group_2'].value_counts()
```

```python
# again, these missing values are not indicative of errors in the data set, some pokemon only have one egg group
# fill in the missing values and round out the data set
raw_df['Egg_Group_2'].fillna('none', inplace = True)
```

```python
# check for missing values again
raw_df['Egg_Group_2'].isnull().any()
```

```python
# check info one last time for missing values
raw_df.info()
```

# Categorical Feature Encoding

```python
# converting isLegendary, hasGender, and hasMegaEvolution to binary
raw_df['isLegendary'] = raw_df['isLegendary'].astype(int)

raw_df['hasGender'] = raw_df['hasGender'].astype(int)

raw_df['hasMegaEvolution'] = raw_df['hasMegaEvolution'].astype(int)
```

```python
# giving dummy values to the essential object columns
raw_df = pd.get_dummies(raw_df, columns = ['Type_1', 'Type_2', 'Color', 'Egg_Group_1', 'Egg_Group_2', 'Body_Style'])
```

```python
# drop the name column
raw_df.drop('Name', axis = 1, inplace = True)
```

```python
# one last change - it would not make sense to include the individual stat values as a predictor of the total stat value
raw_df.drop(['HP', 'Attack', 'Defense', 'Sp_Atk', 'Sp_Def', 'Speed'], axis = 1, inplace = True)
```

```python
# reorder columns
columns = [column for column in raw_df.columns if column != 'Total']
columns = ['Total'] + columns
raw_df = raw_df[columns]
```

```python
# check info
raw_df.info()
```
