# (Work in Progress) Exploring the Raw Data Set

```python
# imports
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
# set the path of the raw data
raw_data_path = os.path.join(os.path.pardir, 'Data_Projects', 'pokemon-raw.csv')
```

```python
# read in the csv file
raw_df = pd.read_csv(raw_data_path)
```

```python
# confirm the type
type(raw_df)
```

```python
# check info
raw_df.info()
```

```python
# view the first five rows
raw_df.head()
```

```python
# view some basic descriptive statistics
raw_df.describe(include = 'all')
```

```python
# how many Pokemon are there in each primary type classification?
raw_df.Type_1.value_counts()
```

```python
# visualize the counts of Pokemon by their primary type classification
%matplotlib inline
raw_df.Type_1.value_counts().plot(kind = 'bar', 
                                  rot = 45, 
                                  title = 'Count of Pokemon by Primary Type', 
                                  color = 'm', 
                                  edgecolor = 'black')
```

```python
# who are the only three pokemon with flying as their primary type in this data set?
raw_df[raw_df['Type_1'] == 'Flying']
```

```python
# What does the distribution of Pokemon by their total stat values look like?
raw_df.Total.plot(kind = 'hist', 
                  title = 'Histogram for Total Stats',
                  color = 'm',
                  edgecolor = 'black',
                  bins = 15)
```

```python
# which types have the most powerful Pokemon?
raw_df[raw_df.Total >= 650].Type_1.value_counts()
```

```python
# which types have the least powerful Pokemon?
raw_df[raw_df.Total <= 250].Type_1.value_counts()
```

```python
# create a table to check the strength of all types
raw_df.groupby('Type_1').Total.mean()
```

```python
# compare histograms of strength of Dragon vs. Bug types
f , (ax1, ax2) = plt.subplots(1, 2 , figsize=(14,3))

ax1.hist(raw_df[raw_df.Type_1 == "Dragon"].Total, 
         range = (0, 750), 
         bins=20, 
         color='c',
         edgecolor = 'black')
ax1.set_title('Histogram of Dragon Type Total Stats')
ax1.set_xlabel('Bins')
ax1.set_ylabel('Counts')

ax2.hist(raw_df[raw_df.Type_1 == "Bug"].Total, 
         range = (0, 750), 
         bins=20, 
         color='m', 
         edgecolor = 'black')
ax2.set_title('Histogram of Bug Type Total Stats')
ax2.set_xlabel('Bins')
ax2.set_ylabel('Counts')

plt.show()
```

```python
# compare catch rate vs. total stats (Note: higher catch rate means a Pokemon is easier to catch)
raw_df.plot.scatter(x = 'Catch_Rate', 
                    y = 'Total', 
                    title = 'Scatter Plot of Catch Rate vs. Total Stats',
                    color = 'm',
                    alpha = 0.2)
```

# Data Munging: Filling Missing Values

```python
# check for missing values
raw_df.info()
```

```python
# start by looking at missing Type_2 values
raw_df[raw_df['Type_2'].isnull()]

# from my own knowledge, I can tell these values are not errors, these pokemon only have one type
```

```python
# fill in the nulls
raw_df['Type_2'].fillna('none', inplace = True)
```

```python
# check to confirm there are no longer any nulls
raw_df.Type_2.isnull().any()
```

```python
# look at the info again
raw_df.info()
```

```python
# next look at the nulls for Pr_Male
raw_df[raw_df['Pr_Male'].isnull()]
```

```python
# notice these nulls indicate a pokemon does not have a gender, so lets fill in the rows with that information
raw_df['Pr_Male'].fillna('No Gender', inplace = True)
```

```python
# check to confirm there are no longer any nulls
raw_df['Pr_Male'].isnull().any()
```

```python
# look at the info again
raw_df.info()
```

```python
# finally, look at the missing values for Egg_Group_2
raw_df[raw_df['Egg_Group_2'].isnull()]
```

```python
# dig a little deeper to get a better understanding of the Egg Groups
raw_df['Egg_Group_1'].value_counts()
```

```python
# now look at Egg_Group_2 groups
raw_df['Egg_Group_2'].value_counts()
```

```python
# again, these missing values are not indicative of errors in the data set, some pokemon only have one egg group
# fill in the missing values and round out the data set
raw_df['Egg_Group_2'].fillna('none', inplace = True)
```

```python
# check for missing values again
raw_df['Egg_Group_2'].isnull().any()
```

```python
# check info one last time for missing values
raw_df.info()
```
